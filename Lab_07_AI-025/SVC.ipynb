```python
import pandas as pd
import numpy as np
from sklearn.svm import SVC
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
data=pd.read_csv("ads.csv")
```

```python
data.head()
```

```python
le = LabelEncoder()
for col in data.columns:
    if data[col].dtype == 'object':
        data[col] = le.fit_transform(data[col])
```

```python
features = ['Daily Time Spent on Site', 'Age', 'Area Income', 'Daily Internet Usage']
target = 'Clicked on Ad'

X = data[features]
y = data[target]
```

```python
sns.pairplot(data, hue='Clicked on Ad', vars=features)
plt.show()

# 3D Visualization
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
```

```python
x = data['Daily Time Spent on Site']
y_3d = data['Age']
z = data['Area Income']
color = data['Clicked on Ad']

sc = ax.scatter(x, y_3d, z, c=color, cmap='coolwarm')
```

```python
ax.set_xlabel('Daily Time Spent on Site')
ax.set_ylabel('Age')
ax.set_zlabel('Area Income')
ax.set_title('3D Visualization of Ad Click Behavior')

# Use fig.colorbar to avoid warning
fig.colorbar(sc, ax=ax, label='Clicked on Ad (0=No, 1=Yes)')

plt.show()
```
